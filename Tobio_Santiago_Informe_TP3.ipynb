```python
def entropia_cruzada(p, q):
    return -np.sum(p * np.log(q))

moneda_cargada = np.array([0.0, 1])
moneda_justa = np.array([0.5, 0.5]) 

print(entropia_cruzada(moneda_cargada, moneda_justa))
print(entropia_cruzada(moneda_justa, moneda_cargada))


# Probabilidades para un dado de 6 caras, una cara con 0.99 y el resto suma 0.01
Q = np.full(6, 0.01 / 5)
Q[0] = 0.99  # Por ejemplo, la primera cara tiene probabilidad 0.99
print(Q)
print("Suma:", Q.sum())
P = np.full(6, 1/6)

# d(x,y) >= 0
print(entropia_cruzada(Q, P))
print(entropia_cruzada(P, Q))
```
